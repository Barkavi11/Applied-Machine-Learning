# Implementation of Simple linear regression, Multiple linear regression, Polynomial regression and Logistic Regression

Performing simple linear regression, multiple linear regression, and polynomial regression to analyze the relationship between engine volume, weight, and CO2 emissions in a cars dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
import statsmodels.api as sm
from sklearn.pipeline import make_pipeline
```

```python
# To set a consistent style for all plots
plt.style.use('seaborn-v0_8')
```

```python
# To load the CSV file correctly
df_raw = pd.read_csv('cars.csv', header=None)
headers = df_raw.iloc[0].tolist()
df = pd.DataFrame(df_raw.values[1:], columns=headers)

# To convert numeric columns to appropriate types
df['Volume'] = pd.to_numeric(df['Volume'])
df['Weight'] = pd.to_numeric(df['Weight'])
df['CO2'] = pd.to_numeric(df['CO2'])

# To display the first few rows of the cleaned dataset
print("Cleaned dataset:")
print(df.head())
```

## Simple Linear Regression (predicts CO2 emissions based solely on engine volume : CO2 = 0.0113  Volume + 83.7464)

```python
# In this case, engine volume is used as the predictor and CO2 emissions as the target
X_simple = df[['Volume']]
y = df['CO2']
```

```python
# To split the data into training and testing sets (80% train, 20% test)
X_train_simple, X_test_simple, y_train, y_test = train_test_split(
    X_simple, y, test_size=0.2, random_state=42)
```

```python
# To create and train the simple linear regression model
simple_model = LinearRegression()
simple_model.fit(X_train_simple, y_train)
```

```python
# To make predictions on the test set
y_pred_simple = simple_model.predict(X_test_simple)
```

```python
# To calculate performance metrics
mse_simple = mean_squared_error(y_test, y_pred_simple)
r2_simple = r2_score(y_test, y_pred_simple)
```

```python
# To print the model coefficients and performance metrics
print(f"Simple Linear Regression Equation: CO2 = {simple_model.coef_[0]:.4f}  Volume + {simple_model.intercept_:.4f}")
print(f"Mean Squared Error: {mse_simple:.4f}")
print(f"R Score: {r2_simple:.4f}")
```

```python
# To visualize the simple linear regression model
plt.figure(figsize=(10, 6))
plt.scatter(X_simple, y, color='blue', alpha=0.7, label='Actual data points')
plt.plot(X_simple, simple_model.predict(X_simple), color='red', linewidth=2, label='Regression line')
plt.title('Simple Linear Regression: Engine Volume vs CO2 Emissions', fontsize=14)
plt.xlabel('Engine Volume (cc)', fontsize=12)
plt.ylabel('CO2 Emissions (g/km)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# To use statsmodels for detailed statistics
X_sm_simple = sm.add_constant(X_simple)  # Add a constant term for the intercept
model_sm_simple = sm.OLS(y, X_sm_simple).fit()
print("Detailed Statistics for Simple Linear Regression:")
print(model_sm_simple.summary())
```

By using engine volume to predict CO2 emissions:

*   Simple Linear Regression Equation: CO2 = 0.0085  Volume + 87.7383
*   Mean Squared Error: 57.0164
*   R Score: 0.3417

## Multiple Linear Regression (predicts CO2 emissions using both engine volume and weight : CO2 = 0.0078  Volume + 0.0076  Weight + 79.6947)

```python
# In this case, both engine volume and weight has been used as predictors
X_multiple = df[['Volume', 'Weight']]
```

```python
# To split the data into training and testing sets
X_train_multiple, X_test_multiple, y_train, y_test = train_test_split(
    X_multiple, y, test_size=0.2, random_state=42)
```

```python
# To create and train the multiple linear regression model
multiple_model = LinearRegression()
multiple_model.fit(X_train_multiple, y_train)
```

```python
# To make predictions on the test set
y_pred_multiple = multiple_model.predict(X_test_multiple)
```

```python
# To calculate performance metrics
mse_multiple = mean_squared_error(y_test, y_pred_multiple)
r2_multiple = r2_score(y_test, y_pred_multiple)
```

```python
# To print the model coefficients and performance metrics
print(f"Multiple Linear Regression Equation: CO2 = {multiple_model.coef_[0]:.4f}  Volume + {multiple_model.coef_[1]:.4f}  Weight + {multiple_model.intercept_:.4f}")
print(f"Mean Squared Error: {mse_multiple:.4f}")
print(f"R Score: {r2_multiple:.4f}")
```

```python
# To create a 3D plot to visualize the multiple regression
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
```

```python
# To create a meshgrid for the prediction surface
x_surf = np.linspace(df['Volume'].min(), df['Volume'].max(), 100)
y_surf = np.linspace(df['Weight'].min(), df['Weight'].max(), 100)
x_surf, y_surf = np.meshgrid(x_surf, y_surf)
z_surf = multiple_model.intercept_ + multiple_model.coef_[0] * x_surf + multiple_model.coef_[1] * y_surf
```

```python
# To plot the surface
ax.plot_surface(x_surf, y_surf, z_surf, alpha=0.5, cmap='viridis')
```

```python
# To plot the actual data points
ax.scatter(df['Volume'], df['Weight'], df['CO2'], c='red', marker='o', alpha=0.8)
```

```python
ax.set_xlabel('Engine Volume (cc)', fontsize=12)
ax.set_ylabel('Weight (kg)', fontsize=12)
ax.set_zlabel('CO2 Emissions (g/km)', fontsize=12)
ax.set_title('Multiple Linear Regression: Volume and Weight vs CO2 Emissions', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
import statsmodels.api as sm
from sklearn.pipeline import make_pipeline
```

```python
df_raw = pd.read_csv('cars.csv', header=None)
headers = df_raw.iloc[0].tolist()
df = pd.DataFrame(df_raw.values[1:], columns=headers)

df['Volume'] = pd.to_numeric(df['Volume'])
df['Weight'] = pd.to_numeric(df['Weight'])
df['CO2'] = pd.to_numeric(df['CO2'])
```

```python
# In this case, engine volume has been used as the predictor with polynomial features
X_simple = df[['Volume']]
y = df['CO2']
```

```python
# To split the data into training and testing sets
X_train_simple, X_test_simple, y_train, y_test = train_test_split(
    X_simple, y, test_size=0.2, random_state=42)

# For different polynomial degrees (try)
degrees = [2, 3]
models_poly = []
mse_poly = []
r2_poly = []

for degree in degrees:
    # To create polynomial features
    poly_features = PolynomialFeatures(degree=degree)
    X_train_poly = poly_features.fit_transform(X_train_simple)
    X_test_poly = poly_features.transform(X_test_simple)

    # To create and train the polynomial regression model
    poly_model = LinearRegression()
    poly_model.fit(X_train_poly, y_train)
    models_poly.append(poly_model)

    # To make predictions on the test set
    y_pred_poly = poly_model.predict(X_test_poly)

    # To calculate performance metrics
    mse = mean_squared_error(y_test, y_pred_poly)
    r2 = r2_score(y_test, y_pred_poly)

    mse_poly.append(mse)
    r2_poly.append(r2)

    print(f"Polynomial Regression (Degree {degree}):")
    print(f"Mean Squared Error: {mse:.4f}")
    print(f"R Score: {r2:.4f}")
```

```python
# By using statsmodels for detailed statistics
X_poly_sm = poly_features.fit_transform(X_simple)
X_poly_sm = sm.add_constant(X_poly_sm)  # Add a constant term for the intercept
model_sm_poly = sm.OLS(y, X_poly_sm).fit()
print(f"Detailed Statistics for Polynomial Regression (Degree {degree}):")
print(model_sm_poly.summary())
```

## Logistic Regression (predicts whether a car has high/low CO2 emissions, binary classification)

```python
# To perform Logistic Regression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc
from sklearn.preprocessing import StandardScaler
df = pd.read_csv('cars.csv')
```

```python
# To create binary target variable
co2_median = df['CO2'].median()
df['HighCO2'] = (df['CO2'] > co2_median).astype(int)
```

```python
# To select features and target
X = df[['Volume', 'Weight']]
y = df['HighCO2']
```

```python
# To split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# To standardize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# To fit logistic regression model
model = LogisticRegression(random_state=42)
model.fit(X_train_scaled, y_train)
```

```python
# To make predictions
y_pred = model.predict(X_test_scaled)
y_pred_prob = model.predict_proba(X_test_scaled)[:, 1]
```

```python
# For the model evaluation
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)
```

```python
print("Logistic Regression Results:")
print(f"Accuracy: {accuracy:.4f}")
print("Confusion Matrix:")
print(conf_matrix)
```

```python
print("Classification Report:")
print(class_report)
```

```python
# To get coefficients
coef = model.coef_[0]
intercept = model.intercept_[0]
print("Model Coefficients:")
print(f"Volume: {coef[0]:.4f}")
print(f"Weight: {coef[1]:.4f}")
print(f"Intercept: {intercept:.4f}")
```

```python
# To calculate odds ratios
odds_ratios = np.exp(coef)
print("Odds Ratios:")
print(f"Volume: {odds_ratios[0]:.4f}")
print(f"Weight: {odds_ratios[1]:.4f}")
```

## Comparison of Polynomial Regression and Logistic Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, classification_report, confusion_matrix, roc_curve, auc
from sklearn.pipeline import Pipeline
df = pd.read_csv('cars.csv')
```

```python
co2_median = df['CO2'].median()
df['HighCO2'] = (df['CO2'] > co2_median).astype(int)

X = df[['Volume', 'Weight']]
y_continuous = df['CO2']
y_binary = df['HighCO2']

X_train, X_test, y_cont_train, y_cont_test, y_bin_train, y_bin_test = train_test_split(
    X, y_continuous, y_binary, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
poly = PolynomialFeatures(degree=2)
X_poly_train = poly.fit_transform(X_train_scaled)
X_poly_test = poly.transform(X_test_scaled)

poly_model = LinearRegression()
poly_model.fit(X_poly_train, y_cont_train)

y_poly_pred = poly_model.predict(X_poly_test)

poly_mse = mean_squared_error(y_cont_test, y_poly_pred)
poly_r2 = r2_score(y_cont_test, y_poly_pred)
```

```python
log_model = LogisticRegression(random_state=42)
log_model.fit(X_train_scaled, y_bin_train)

y_log_pred = log_model.predict(X_test_scaled)
y_log_prob = log_model.predict_proba(X_test_scaled)[:, 1]

log_accuracy = accuracy_score(y_bin_test, y_log_pred)
log_conf_matrix = confusion_matrix(y_bin_test, y_log_pred)
log_report = classification_report(y_bin_test, y_log_pred, output_dict=True)

fpr, tpr, _ = roc_curve(y_bin_test, y_log_prob)
log_auc = auc(fpr, tpr)
```

```python
# Cross-validation for more robust comparison
# For polynomial regression
poly_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('poly', PolynomialFeatures(degree=2)),
    ('regression', LinearRegression())
])
poly_cv_scores = cross_val_score(poly_pipeline, X, y_continuous, cv=5, scoring='r2')
```

```python
# For logistic regression
log_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('logistic', LogisticRegression(random_state=42))
])
log_cv_scores = cross_val_score(log_pipeline, X, y_binary, cv=5, scoring='accuracy')
```
